```python
import os
from pathlib import Path
import string
import pandas as pd
from supabase import create_client, Client

from dotenv import load_dotenv
load_dotenv()

url = os.getenv('SUPABASE_URL_TP')
key = os.getenv('SUPABASE_KEY_TP')
supabase = create_client(supabase_url=url, supabase_key=key)
```

```python
response = supabase.table('subsections').select("*").execute()
subsections_df = pd.DataFrame(response.data)
subsections_df
```

```python
kb_dir = Path('../kb').resolve()

for row in subsections_df.itertuples():
    p = (kb_dir / row.subsection_id).with_suffix('.txt')
    if row.clean_text:
        p.write_text(row.clean_text)
```
